```python
df["helpful_no"] = df["total_vote"] - df["helpful_yes"]
```

```python
def score_up_down_diff(up, down):
    return up - down
```

```python
df["score_up_down_diff"] = df.apply(lambda x: score_up_down_diff(x["helpful_yes"],x["helpful_no"]), axis=1)
```

```python
def score_average_rating(up, down):
    if up + down == 0:
        return 0
    return up / (up + down)
```

```python
df["score_average_rating"] = df.apply(lambda x: score_average_rating(x["helpful_yes"],x["helpful_no"]), axis=1)
```

```python
import math
import scipy.stats as st
```

```python
def wilson_lower_bound(up, down, confidence=0.95):
    """
    Wilson Lower Bound Score hesapla

    - Bernoulli parametresi p iÃ§in hesaplanacak gÃ¼ven aralÄ±ÄŸÄ±nÄ±n alt sÄ±nÄ±rÄ± WLB skoru olarak kabul edilir.
    - Hesaplanacak skor Ã¼rÃ¼n sÄ±ralamasÄ± iÃ§in kullanÄ±lÄ±r.
    - Not:
    EÄŸer skorlar 1-5 arasÄ±daysa 1-3 negatif, 4-5 pozitif olarak iÅŸaretlenir ve bernoulli'ye uygun hale getirilebilir.
    Bu beraberinde bazÄ± problemleri de getirir. Bu sebeple bayesian average rating yapmak gerekir.

    Parameters
    ----------
    up: int
        up count
    down: int
        down count
    confidence: float
        confidence

    Returns
    -------
    wilson score: float

    """
    n = up + down
    if n == 0:
        return 0
    z = st.norm.ppf(1 - (1 - confidence) / 2)
    phat = 1.0 * up / n
    return (phat + z * z / (2 * n) - z * math.sqrt((phat * (1 - phat) + z * z / (4 * n)) / n)) / (1 + z * z / n)
```

```python
df["wilson_lower_bound_ratings"] = df.apply(lambda x: wilson_lower_bound(x["helpful_yes"],x["helpful_no"]), axis=1)
```

```python
(df.sort_values( by = "wilson_lower_bound_ratings" , ascending = False).head(20)).loc[:,["helpful_yes","helpful_no","score_up_down_diff","score_average_rating","wilson_lower_bound_ratings"]]
```

Different rankings were formed by all three methods. In the up-down technique, the up percentage is ignored, and the total number of votes given in the average rating method is ignored. The most logical results were obtained with the Wilson lower bound method.
